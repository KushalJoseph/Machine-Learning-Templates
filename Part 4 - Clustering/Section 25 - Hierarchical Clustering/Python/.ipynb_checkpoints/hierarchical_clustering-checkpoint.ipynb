# Hierarchical Clustering

## Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```python
dataset = pd.read_csv('Mall_Customers.csv')
X = dataset.iloc[:, [3, 4]].values
```

## Using the dendrogram to find the optimal number of clusters

```python
import scipy.cluster.hierarchy as sch
dendrogram = sch.dendrogram(sch.linkage(X, method='ward'))
plt.title("Dendrogram")
plt.xlabel("Customers")
plt.ylabel("Distances between clusters")
plt.show()
```

## Training the Hierarchical Clustering model on the dataset

```python
from sklearn.cluster import AgglomerativeClustering as ac
hc = ac(n_clusters = 6, affinity = 'euclidean', linkage = 'ward') # "Ward": Minimum Variance
dependent_variable = hc.fit_predict(X)
# Fit just fits it. Fit-pred returns, for each data row, which cluster number it is in
print(dependent_variable)
```

## Visualising the clusters

```python
colors = {"0":"red","1":"blue","2":"green","3":"cyan","4":"yellow","5":"pink","6":"orange"}
for i in range (0,6):
    plt.scatter(X[dependent_variable == i, 0], X[dependent_variable == i,1], s = 70, c = colors[str(i)], label = f"Cluster {i+1}")

plt.title("Clusters")
plt.xlabel("Annual Income")
plt.ylabel("Spending Index")
# plt.legend()
plt.show()
```
